####Converting the columns having datetime format into Timestamp columns

```python
data['Extinguished'] = pd.to_datetime(pd.to_datetime(data['Extinguished']).dt.strftime('%Y-%m-%d'), format='%Y-%m-%d')
data['Started'] = pd.to_datetime(pd.to_datetime(data['Started']).dt.strftime('%Y-%m-%d'), format='%Y-%m-%d')
```

####Filling in the missing values of Extinguished

```python
from datetime import datetime
df = data[(data['Started'].notnull() & data['Extinguished'].notnull())]
fire_active_days = df['Extinguished'] - df['Started']
fire_active_days.head()
```

```python
fire_active_days = fire_active_days.apply(lambda x: x.days)
```

```python
plt.figure(figsize = (30,10))
fire_active_days.value_counts().plot.bar(color ='green',  width = 0.4) 
plt.show()
```

```python
from datetime import timedelta
data.loc[data['Extinguished'].isnull(),'Extinguished'] = data.loc[data['Extinguished'].isnull(),'Started'].apply(lambda x: x + timedelta(days=1))
```

```python
data['Extinguished'].isnull().sum()
```

```python
data['AcresBurned'].mean()
```

```python
data['AcresBurned'].fillna(data['AcresBurned'].mean(), inplace=True)
```

####Filling in the missing values of PercentContained

```python
data['PercentContained'].fillna(100.0, inplace=True)
```

```python
import folium
from folium.plugins import HeatMap
from IPython.display import HTML

# interactive map for visualizing
zoom_factor = 5 # inital map size
radius_scaling = 50 # scaling of bubbles

for year in sorted(data['YearFireStarted'].unique()):
    my_map = folium.Map(location=[36,-120], zoom_start=zoom_factor)
    df = data[(data['YearFireStarted'] == int(year))]
    print(str(df.shape[0])+" fires in "+ str(year))
    counties = pd.DataFrame(df['Counties'].value_counts())
    counties['index'] = counties.index
    for i in range(0, counties.shape[0]):
        small_df = df[df['Counties'] == counties.iloc[i]['index']]
        folium.Circle(
            location=[small_df.iloc[0]['Latitude'], small_df.iloc[0]['Longitude']],
            radius=np.sqrt(small_df['AcresBurned'].sum())*radius_scaling,
            color='orange',
            popup='County:' +counties.iloc[i]['index']+  ' - No. of fires:'+str(small_df.shape[0]) +' - Acres Burned:' +str(small_df['AcresBurned'].sum()),
            fill=True,
            fill_color='red'
        ).add_to(my_map)

    display(my_map)
```

```python
zoom_factor = 4 # inital map size
radius_scaling = 10 # scaling of bubbles

fire_per_location = {}
for county in sorted(data['Counties'].unique()):
    
    df = data[(data['Counties'] == county)]
    if df['YearFireStarted'].nunique() >= 7:
        fire_per_location[county] = {'count': df.shape[0],'AcresBurned':df['AcresBurned'].sum() ,'latitude': df['Latitude'].iloc[0], 'longitude': df['Longitude'].iloc[0], 'noOfYears': df['YearFireStarted'].nunique()}

map = folium.Map(location=[36,-120], zoom_start=zoom_factor)
for county in fire_per_location.keys():
    folium.Circle(
        location=[fire_per_location[county]['latitude'], fire_per_location[county]['longitude']],
        radius=np.sqrt(fire_per_location[county]['AcresBurned'].sum()*fire_per_location[county]['count'])*radius_scaling,
        color='red',
        popup='No. of times fire occured in '+ county +' in last '+ str(fire_per_location[county]['noOfYears'])+ ' years is '+ str(fire_per_location[county]['count']) + ' burning approx '+ str(round(fire_per_location[county]['AcresBurned'].sum(),2)) + ' acres',
        fill=True,
        fill_color='red'
    ).add_to(map)
map
```
